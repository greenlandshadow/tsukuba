```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 18
```

```python
def generate(t):
    a = 0.5
    x = x_1 =1.0
    xs =[x_1]
    for i in range(1,t):
        x= a*x
        xs.append(x)
    return np.array(xs)

xs= generate(20)
ts=np.arange(20)
plt.xlabel("discrete time($t$)")
plt.plot(ts,xs,"o",label="iteration")
plt.plot(ts,0.5**ts,label="theory")
plt.legend()
plt.show()
```
