```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 18
```

```python
def generate(t):
    a_1,a_2,c = 1.5,-2.0,1.0
    x,y = x_2,x_1 =1.0,2.0
    xs =[x_1,x_2]
    for i in range(2,t):
        x_temp,y_temp=x,y
        x= a_1*x_temp +a_2*y_temp+c
        y=x_temp
        xs.append(x)
    return np.array(xs)

xs= generate(20)
ts=np.arange(20)
plt.xlabel("discrete time($t$)")
plt.plot(ts,xs,"o",label="iteration")
#plt.plot(ts,0.5**ts,label="theory")
plt.legend()
plt.show()
```
