```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_steps = 50


y_initial = [0, 0]

def compute_yt_2_1_2(t_steps, y_initial):
    y = np.zeros(t_steps)
    y[:2] = y_initial
    for t in range(2, t_steps):
        y[t] = 2 * y[t-1] + y[t-2] + 2
    return y


y_2_1_2 = compute_yt_2_1_2(t_steps, y_initial)


plt.figure(figsize=(12, 8))
plt.plot(np.arange(t_steps), y_2_1_2, label='y_{t+1} = 2y_t + y_{t-1} + 2')
plt.xlabel('t')
plt.ylabel('y_t')
plt.legend()
plt.grid(True)
plt.show()
```
