```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dx_dt(t, x):
    return -x + np.sin(t)
x0 = 0
t0 = 0
t_end = 10
h = 0.1
t_points = np.arange(t0, t_end + h, h)
x_points = np.zeros(len(t_points))
x_points[0] = x0

for i in range(1, len(t_points)):
    x_points[i] = x_points[i-1] + h * dx_dt(t_points[i-1], x_points[i-1])

plt.figure(figsize=(10, 6))
plt.plot(t_points, x_points, label='Euler Method', color='blue')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()
```
