```python
import numpy as np
import matplotlib.pyplot as plt

def erlang_b(rho, c):
    if c == 0:
        return 1.0
    else:
        prev = erlang_b(rho, c-1)
        return (rho * prev) / (c + rho * prev)

rhos = np.linspace(0, 100, 500)
servers_list = [5, 10, 15, 20, 25]

plt.figure(figsize=(12, 6))

for servers in servers_list:
    blocking_probabilities = [erlang_b(rho, servers) for rho in rhos]
    plt.plot(rhos, blocking_probabilities, label=f'C = {servers}')

plt.title('Erlang B Blocking Probability')
plt.xlabel('ρ (Erlang)')
plt.ylabel('Blocking probability')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
